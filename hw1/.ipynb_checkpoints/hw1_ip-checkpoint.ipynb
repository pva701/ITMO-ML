```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pnd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cross_validation import KFold
from itertools import product
```

```python
arr = np.genfromtxt('data/chips.csv', delimiter=',')
xs, ys = arr[:,:2].copy(), np.int32(arr[:, 2:].ravel())
```

```python
def show(xs, ys):
    label0, = plt.plot(xs[ys==0][:, 0], xs[ys==0][:, 1], 'bo')
    label1, = plt.plot(xs[ys==1][:, 0], xs[ys==1][:, 1], 'ro')
    plt.legend([label0, label1], ['0', '1'])
    plt.show()
show(xs, ys)

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(xs)
show(x_scaled, ys)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(xs)
show(x_scaled, ys)
```

```python
# http://stats.stackexchange.com/questions/44166/kernelised-k-nearest-neighbour
import sklearn.metrics.pairwise as pw

def knn(x, data, labels, k, weight_f='uniform', metric_f='l2', transform_f='id', scaler=None, **params):
    if transform_f == 'id':
        transform_f = lambda x: x
    if weight_f == 'uniform':
        weight_f = lambda x: x
    if type(metric_f) is str:
        metric = metric_f
        metric_f = lambda x, y: pw.pairwise_distances([x, y], metric=metric, **params)[0, 1]
        #metric_f = lambda x, y: ((x-y)**2).sum()**0.5
    if scaler is not None:
        x = scaler.transform(x.reshape(1, -1)).ravel()
        data = scaler.transform(data)
        
    mapped_x = transform_f(x)
    f = lambda t: weight_f(metric_f(mapped_x, transform_f(t)))
    arr = np.apply_along_axis(f, 1, data)
    neighbors = np.argsort(arr)[:k]
    neighbors_labels = labels[neighbors]
    x_label = np.bincount(neighbors_labels).argmax()
    return x_label
```

```python
def f1_score(predicted, true):
    tp = ((predicted == true) * (predicted == 1)).sum()
    fp = ((predicted != true) * (predicted == 1)).sum()
    fn = ((predicted != true) * (predicted == 0)).sum()
    p = tp*1.0/(tp+fp)
    r = tp*1.0/(tp+fn)
    return 2*p*r/(p+r)
    

def print_args(func):
    def func_wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        to_str = lambda x: x.__name__ if callable(x) else str(x)
        print("Parameters: ", ",".join([str(key) + "=" + to_str(value) for key, value in kwargs.items()]))
        print("Result: ", str(result))
        return result
    return func_wrapper

@print_args
def kf_cross_validation(xs, ys, n_fold=10, measure='accuracy', **params):
    if measure == 'accuracy':
        measure = lambda x, y: (x==y).sum() * 1.0/len(x)
    kf = KFold(len(xs), n_fold, True)
    kf_sum = 0
    for train_i, test_i in kf:
        classifier = lambda x: knn(x, xs[train_i], ys[train_i], **params)
        true_labels = ys[test_i]
        predicted_labels = np.apply_along_axis(classifier, 1, xs[test_i])
        kf_sum += measure(predicted_labels, true_labels)
    return kf_sum/n_fold
```

```python
def transform1(x):
    return np.array([x[0], x[1], 2*x[0]*x[1], x[0]**2, x[1]**2])

def transform2(x):
    return np.array([x[0], x[1], x[0]**2, x[1]**2, x[0]**3, x[1]**3])
```

```python
std_scaler = StandardScaler().fit(xs)
n_fold = 10
kf_cross_validation(xs, ys, n_fold, k=5)
kf_cross_validation(xs, ys, n_fold, k=5, transform_f=transform1)
kf_cross_validation(xs, ys, n_fold, k=3, transform_f=transform1)
kf_cross_validation(xs, ys, n_fold, k=5, metric_f='manhattan')
kf_cross_validation(xs, ys, n_fold, k=5, metric_f='manhattan', transform_f=transform2)
kf_cross_validation(xs, ys, n_fold, k=3, metric_f='manhattan', transform_f=transform1)
kf_cross_validation(xs, ys, n_fold, k=3, transform_f=transform1, metric_f='minkowski', p=3)
kf_cross_validation(xs, ys, n_fold, k=5, transform_f=transform1, metric_f='minkowski', p=3)
kf_cross_validation(xs, ys, n_fold, k=7, transform_f=transform1, metric_f='minkowski', p=3)
```

```python
x_min, x_max = xs[:, 0].min() - 0.3, xs[:, 0].max() + 0.3
y_min, y_max = xs[:, 1].min() - 0.3, xs[:, 1].max() + 0.3
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
plt.figure(figsize=(10,10))
classifier = lambda x: knn(x, xs, ys, k=5, weight_f=lambda x: x**2)          
Z = np.apply_along_axis(classifier, 1, np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.1)
plt.scatter(xs[:, 0], xs[:, 1], s=50, c=ys)
plt.show()
```
